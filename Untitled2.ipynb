```python
from transformers import pipeline
```

```python
pipe = pipeline(
    "text-generation",
    model="microsoft/Phi-3-mini-4k-instruct",
    tokenizer="microsoft/Phi-3-mini-4k-instruct",
    dtype="auto",
    device_map="auto",
    max_new_tokens=120,
    max_length=None
)
```

```python
### Single‑Report Simplification Function
```

```python
def simplify_hematology(text):
    prompt = f"Simplify this hematology lab report into plain English:\n\n{text}\n\nExplanation:"
    output = pipe(prompt)[0]["generated_text"]
    return output
```

```python
### 4. Batch Processing Function
```

```python
def batch_simplify(df, text_column="report"):
    simplified = []
    for text in df[text_column]:
        explanation = simplify_hematology(text)
        simplified.append(explanation)
    df["simplified"] = simplified
    return df
```
