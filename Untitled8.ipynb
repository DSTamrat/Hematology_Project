```python
### 5-  Building a simple rule‑based “simplifier”  as a baseline  

### Before using Generative AI, we’ll create a very simple glossary‑based simplifier.
### This gives us a baseline and makes the LLM’s value obvious.
```

```python
medical_to_lay = {
    "leukocytosis": "a higher than normal number of white blood cells",
    "neutrophilia": "an increase in a type of white blood cell that fights bacteria",
    "anemia": "a lower than normal number of red blood cells",
    "hemoglobin": "the protein in red blood cells that carries oxygen",
    "hematocrit": "the percentage of blood made up of red blood cells",
    "microcytic": "smaller than normal red blood cells",
    "leukopenia": "a lower than normal number of white blood cells",
    "lymphopenia": "a lower than normal number of lymphocytes, a type of white blood cell",
    "thrombocytopenia": "a lower than normal number of platelets",
    "platelets": "cells that help your blood clot and stop bleeding",
    "CBC": "a complete blood count (CBC), a common blood test",
}
```

```python
### Explanation:
### medical_to_lay – maps medical terms to short explanations.
### simple_glossary_simplifier – goes through the text and replaces each term with term (explanation)
```

```python
def simple_glossary_simplifier(text, glossary):
    simplified = text
    for term, explanation in glossary.items():
        simplified = simplified.replace(term, f"{term} ({explanation})")
    return simplified
```

### 6 Applying  the baseline simplifier

```python
df["simplified_baseline"] = df["report_text"].apply(
    lambda t: simple_glossary_simplifier(t, medical_to_lay)
)

for _, row in df.iterrows():
    print(f"Report ID: {row['report_id']}")
    print("Original:")
    print(fill(row['report_text'], width=80))
    print("\nBaseline simplified:")
    print(fill(row['simplified_baseline'], width=80))
    print("=" * 80)
```

```python
### In the above, we created a new column, simplified_baseline. For each report, we show: 
### Original text 
### Baseline simplified version with inline explanations.
### This is still not truly patient‑friendly, but it’s a good starting point.
```

### 7 - Add a Generative AI layer (LLM) – structure

```python
### Now we design how we’d call a Generative AI model. 
### I’ll show the structure using a generic call_llm() function—you can plug in your provider (OpenAI, Azure, etc.).

### Explanation:
### system_instructions – sets the role and constraints for the model.
### user_prompt – gives the actual report and asks for a simplified explanation.
### We return both so they can be passed to an LLM API.
```

```python
def build_prompt(report_text):
    system_instructions = (
        "You are a clinical educator. Your job is to rewrite hematology lab reports "
        "so that a non-medical patient can understand them. Preserve the medical "
        "meaning, do not add new diagnoses, and avoid giving medical advice. "
        "Use short sentences and simple language at about an 8th grade reading level."
    )

    user_prompt = (
        "Rewrite the following hematology report in patient-friendly language. "
        "You may explain medical terms, but do not change the meaning.\n\n"
        f"Report:\n{report_text}\n\n"
        "Patient-friendly explanation:"
    )

    return system_instructions, user_prompt
```

```python
def call_llm(text):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You simplify hematology lab reports into plain English."},
            {"role": "user", "content": text}
        ]
    )
    return response.choices[0].message["content"]
```

```python

from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
model_name = "microsoft/Phi-3-mini-4k-instruct"

pipe = pipeline(
    "text-generation",
    model=model_name,
    tokenizer=model_name,
    dtype="auto",
    device_map="auto",
    max_new_tokens=120,
    max_length=None
)
```

```python
def simplify_hematology(text):
    prompt = f"Simplify this hematology lab report into plain English:\n\n{text}\n\nExplanation:"
    result = pipe(prompt)[0]["generated_text"]
    return result
```

```python
import pandas as pd
import random

def generate_random_lab_report():
    # Random but clinically plausible ranges
    wbc = round(random.uniform(3.0, 20.0), 1)          # x10^9/L
    hgb = round(random.uniform(6.0, 18.0), 1)          # g/dL
    plt = random.randint(50, 600)                      # x10^9/L

    # Flags
    if wbc < 4.0:
        wbc_flag = "Low"
    elif wbc > 11.0:
        wbc_flag = "High"
    else:
        wbc_flag = "Normal"

    if hgb < 12.0:
        hgb_flag = "Low"
    elif hgb > 17.0:
        hgb_flag = "High"
    else:
        hgb_flag = "Normal"

    if plt < 150:
        plt_flag = "Low"
    elif plt > 450:
        plt_flag = "High"
    else:
        plt_flag = "Normal"

    report = f"""
WBC: {wbc} ({wbc_flag})
Hemoglobin: {hgb} ({hgb_flag})
Platelets: {plt} ({plt_flag})
"""
    return report.strip()

# Generate 1000 synthetic reports
records = []
for i in range(1, 1001):
    records.append({
        "id": i,
        "report": generate_random_lab_report()
    })

df = pd.DataFrame(records)
df.head()
```

```python
def batch_simplify(df, text_column="report"):
    simplified = []
    
    for text in df[text_column]:
        explanation = simplify_hematology(text)
        simplified.append(explanation)
    
    df["simplified"] = simplified
    return df
```

```python
from tqdm import tqdm

def batch_simplify_fast(df, text_column="report", batch_size=8):
    texts = df[text_column].tolist()
    simplified = []

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]

        prompts = [
            f"Simplify this hematology lab report into plain English:\n\n{t}\n\nExplanation:"
            for t in batch
        ]

        outputs = pipe(prompts)
        simplified.extend([o["generated_text"] for o in outputs])

    df["simplified"] = simplified
    return df
```

```python
from tqdm import tqdm

def batch_simplify_with_progress(df, text_column="report"):
    simplified = []
    
    for text in tqdm(df[text_column], desc="Simplifying reports"):
        explanation = simplify_hematology(text)
        simplified.append(explanation)
    
    df["simplified"] = simplified
    return df
```
